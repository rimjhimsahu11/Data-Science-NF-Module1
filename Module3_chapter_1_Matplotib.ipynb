```python
import matplotlib.pyplot as plt

# define dataset
age = [18,20,23,21,25]
salary = [18000,15000,20000,19000,30000]

#adjust figure size
plt.figure(figsize = (5,10))

#plot line charts
plt.plot(age,salary,marker = '>',linestyle = ':',linewidth = 4,)

##add lebels and title in charts
plt.xlabel("age of employe",fontsize = 12,color = 'purple')
plt.ylabel("Salary of employe",fontsize = 12,color = 'blue')
plt.title('Age vs Salary',fontsize = 14,color = 'green')

#add grid
plt.grid(True)
plt.gca().set_facecolor('pink')

#show the chart
plt.show()
```
